# **Prepare rules for the my book the data sets**

1. Try different values of support and confidence.
   Observe the change in number of rules for different support,confidence values
2. Change the minimum length in apriori algorithm
3. Visulize the obtained rules using different plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing  import TransactionEncoder
```

```python
book = pd.read_csv('/content/book.csv')
book
```

Apriori Algorithm
# 1.Association rules with 10% Support and 70% confidence

```python
from mlxtend import frequent_patterns
frequent_itemsets=apriori(book,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False)
```

```python
rules[rules.lift>1]
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
