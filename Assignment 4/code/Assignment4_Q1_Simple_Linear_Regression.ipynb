**Delivery time - Predict delivery time using sorting time**

Imported Libraries And Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (5,3)
plt.rcParams['figure.dpi'] = 200
```

```python
dataset= pd.read_csv('/content/delivery_time.csv')
dataset.head()
```

# EDA and Data visualization

```python
dataset.info()
```

# Visualizing The Data

```python
fig, ax = plt.subplots(2,2)
plt.tight_layout()

sns.boxplot(dataset['Delivery Time'], ax = ax[0,0],orient = 'h')
sns.boxplot(dataset['Sorting Time'], ax = ax[0,1], orient = 'h')
sns.distplot(dataset['Delivery Time'], ax = ax[1,0])
sns.distplot(dataset['Sorting Time'], ax = ax[1,1])
```

# Feature Engineering

```python
dataset = dataset.rename({'Delivery Time':'delivery_time','Sorting Time':'sorting_time'},axis=1)
dataset
```

# Correlation Analysis

```python
dataset.corr()
```

```python
sns.regplot(x=dataset['sorting_time'],y=dataset['delivery_time'])
```

# Model Building


```python
model = smf.ols('delivery_time~sorting_time', data=dataset).fit()
```

# **Model Tesing**

```python
model.params
```

```python
model.tvalues , model.pvalues
```

```python
model.rsquared , model.rsquared_adj
```

# **Model Predictions**

```python
# Manual prediction for say sorting time 5
delivery_time = (6.582734) + (1.649020)*(5)
delivery_time
```

```python
# Automatic Prediction for say sorting time 5,8

new_data = pd.Series([5,8])
new_data
```

```python
data_pred = pd.DataFrame(new_data, columns=['sorting_time'])
data_pred
```

```python
model.predict(data_pred)
```
